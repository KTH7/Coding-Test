```python
#스택 구현 예제
stack=[]

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop()
stack.append(1)
stack.append(4)
stack.pop()

print(stack)
print(stack[::-1])
```

```python
#큐 구현 예제
from collections import deque
queue=deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) #먼저 들어온 순서
queue.reverse()
print(queue) #나중에 들어온 순서
```

```python
#재귀 함수( 자기 자신을 다시 호출하는 함수 )
def recursive_function(i):
    if i == 100:
        return
    print(i,'번째 재귀함수에서',i+1,'번째 재귀 함수를 호출합니다.')
    recursive_function(i+1)
    print(i,'번째 재귀함수를 종료합니다.')
    
recursive_function(1)
```

```python
#팩토리얼 구현
def factorial_iterative(n):
    result=1
    for i in range(1,n+1):
        result*=i
    return result

def factorial_recursive(n):
    if n<=1:
        return 1
    return n *factorial_recursive(n-1)

print('반복적으로 구현:',factorial_iterative(5))
print('재귀적으로 구현:',factorial_recursive(5))
```

```python
#최대공약수 계산
def GCD(A,B):
    if A%B ==0:
        return B
    else:
        return GCD(B,A%B)
    
print(GCD(192,162))
```

```python
#DFS : 깊이 우선 탐색

def dfs(graph,v,visited):
    visited[v]=True
    print(v,end=' ')
    for i in graph[v]:
        if not visited[i]:
            dfs(graph,i,visited)
            
graph = [ #인접한노드, 0번노드 비워둠
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited=[False]*9

dfs(graph,1,visited)
```

```python
#BFS : 너비 우선 탐색
from collections import deque

def bfs(graph,start,visited):
    queue=deque([start])
    visited[start]=True
    while queue:
        v=queue.popleft()
        print(v,end=' ')
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i]=True
            
graph = [ #인접한노드, 0번노드 비워둠
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited=[False]*9

bfs(graph,1,visited)
```
